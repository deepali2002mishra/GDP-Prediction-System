```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Load datasets
production_df = pd.read_csv(r"D:\Projects\GDP\data\raw\crop_export_production_stable.csv")
climate_df = pd.read_csv(r"D:\Projects\GDP\data\raw\india_climate_soil_1961_2017.csv")

# Merge datasets
merged_df = pd.merge(production_df, climate_df, on=['State', 'Year'], how='left')

# Data cleaning
merged_df.replace([np.inf, -np.inf], np.nan, inplace=True)
merged_df.dropna(subset=['Production Quantity', 'Export Volume', 'Annual Rainfall (mm)'], inplace=True)

def analyze_state(state_name, state_df):
    crop_results = []
    forecast_data = {}
    
    for crop in state_df['Crop'].unique():
        crop_df = state_df[state_df['Crop'] == crop].sort_values('Year')
        
        if len(crop_df) < 5:  # Minimum 5 years of data required
            continue
            
        try:
            # Fit SARIMAX model
            model = SARIMAX(
                endog=crop_df['Production Quantity'],
                exog=crop_df[['Export Volume', 'Annual Rainfall (mm)']],
                order=(1, 1, 1),
                seasonal_order=(0, 0, 0, 0)
            )
            model_fit = model.fit(disp=False)
            
            # Prepare future exogenous variables (using last 3 years average)
            future_exog = pd.DataFrame({
                'Export Volume': [crop_df['Export Volume'].iloc[-3:].mean()] * 10,
                'Annual Rainfall (mm)': [crop_df['Annual Rainfall (mm)'].iloc[-3:].mean()] * 10
            })
            
            # Forecast next 10 years
            forecast = model_fit.get_forecast(steps=10, exog=future_exog)
            forecast_mean = forecast.predicted_mean
            conf_int = forecast.conf_int()
            
            # Calculate metrics
            avg_price = crop_df['Wholesale Price'].mean()
            growth_rate = (forecast_mean.mean() - crop_df['Production Quantity'].iloc[-1]) / crop_df['Production Quantity'].iloc[-1]
            price_volatility = crop_df['Wholesale Price'].std() / crop_df['Wholesale Price'].mean()
            export_dependence = crop_df['Export Volume'].mean() / crop_df['Production Quantity'].mean()
            rainfall_variability = crop_df['Annual Rainfall (mm)'].std() / crop_df['Annual Rainfall (mm)'].mean()
            
            # Soil analysis
            soil_ph = crop_df['Soil pH Level'].mean()
            organic_matter = crop_df['Organic Matter (%)'].mean()
            dominant_soil = crop_df['Soil Type'].mode()[0]
            
            # Soil suitability scoring (0-5 scale)
            soil_score = 0
            if 6 <= soil_ph <= 7:  # Ideal pH range
                soil_score += 3
            elif 5.5 <= soil_ph < 6 or 7 < soil_ph <= 7.5:
                soil_score += 2
                
            if organic_matter >= 2:
                soil_score += 2
            elif 1 <= organic_matter < 2:
                soil_score += 1
                
            # Calculate final score
            base_score = forecast_mean.mean() * avg_price * (1 + growth_rate)
            adjusted_score = base_score * (1 + soil_score/5) * (1 - price_volatility/2)
            
            crop_results.append({
                'State': state_name,
                'Crop': crop,
                'Current_Production': crop_df['Production Quantity'].iloc[-1],
                'Forecasted_Production': forecast_mean.mean(),
                'Growth_Rate': growth_rate,
                'Avg_Price': avg_price,
                'Price_Volatility': price_volatility,
                'Export_Dependence': export_dependence,
                'Rainfall_Variability': rainfall_variability,
                'Soil_pH': soil_ph,
                'Organic_Matter': organic_matter,
                'Soil_Score': soil_score,
                'Score': adjusted_score
            })
            
            # Store forecast data for plotting
            forecast_data[crop] = {
                'history': crop_df['Production Quantity'],
                'forecast': forecast_mean,
                'conf_int': conf_int,
                'years': crop_df['Year'].tolist() + [max(crop_df['Year'])+i+1 for i in range(10)]
            }
            
        except Exception as e:
            continue
    
    if not crop_results:
        return None, None
    
    # Create results dataframe and rank crops
    results_df = pd.DataFrame(crop_results)
    results_df['State_Rank'] = results_df['Score'].rank(ascending=False)
    results_df = results_df.sort_values('State_Rank')
    
    return results_df, forecast_data

def plot_state_forecast(state_name, top_crop, forecast_data):
    plt.figure(figsize=(14, 7))
    data = forecast_data[top_crop['Crop']]
    
    plt.plot(data['years'][:-10], data['history'], 'b-', label='Historical', linewidth=2)
    plt.plot(data['years'][-10:], data['forecast'], 'r--', label='10-Year Forecast', linewidth=2)
    plt.fill_between(data['years'][-10:],
                    data['conf_int'].iloc[:, 0],
                    data['conf_int'].iloc[:, 1],
                    color='pink', alpha=0.3, label='Confidence Interval')
    
    plt.title(f'{top_crop["Crop"]} Production Forecast in {state_name}\n10-Year Investment Score: {top_crop["Score"]:,.2f}', pad=20)
    plt.xlabel('Year', labelpad=10)
    plt.ylabel('Production (tonnes)', labelpad=10)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def print_state_report(state_name, state_results, forecast_data):
    print(f"\n{'='*80}")
    print(f"AGRICULTURAL ANALYSIS FOR {state_name.upper()} (10-YEAR FORECAST)")
    print(f"{'='*80}")
    
    if state_results is None or state_results.empty:
        print("\nNo valid crop data available for this state.")
        return
    
    top_state_crop = state_results.iloc[0]
    
    print(f"\nüèÜ TOP CROP: {top_state_crop['Crop'].upper()}")
    print(f"üìç Current Production: {top_state_crop['Current_Production']:,.2f} tonnes")
    print(f"üîÆ Forecasted Production (10yr avg): {top_state_crop['Forecasted_Production']:,.2f} tonnes")
    print(f"üìà 10-Year Growth Rate: {top_state_crop['Growth_Rate']:.2%}")
    print(f"üí∞ Average Price: ‚Çπ{top_state_crop['Avg_Price']:,.2f}")
    print(f"üå± Soil Score: {top_state_crop['Soil_Score']}/5")
    print(f"üèÜ 10-Year Investment Score: {top_state_crop['Score']:,.2f}")
    
    # Print top 3 crops for the state
    print("\nTOP 3 CROPS IN THIS STATE:")
    top_3 = state_results.head(3)
    print(top_3[['Crop', 'Score', 'Growth_Rate', 'Avg_Price']].to_string(index=False, formatters={
        'Score': '{:,.2f}'.format,
        'Growth_Rate': '{:.2%}'.format,
        'Avg_Price': '‚Çπ{:,.2f}'.format
    }))
    
    # Plot forecast for top crop
    if top_state_crop['Crop'] in forecast_data:
        plot_state_forecast(state_name, top_state_crop, forecast_data)

def generate_investment_rationale(state, crop, merged_df, national_results):
    # Get state-specific data
    state_crop_data = merged_df[(merged_df['State'] == state) & (merged_df['Crop'] == crop)]
    
    # Calculate market share
    national_prod = merged_df[merged_df['Crop'] == crop]['Production Quantity'].sum()
    state_prod = state_crop_data['Production Quantity'].sum()
    market_share = (state_prod / national_prod) * 100 if national_prod > 0 else 0
    
    # Calculate export metrics
    export_ratio = state_crop_data['Export Volume'].mean() / state_crop_data['Production Quantity'].mean() * 100
    
    # Get climate and soil data
    rainfall_var = state_crop_data['Annual Rainfall (mm)'].std() / state_crop_data['Annual Rainfall (mm)'].mean() * 100
    soil_ph = state_crop_data['Soil pH Level'].mean()
    organic_matter = state_crop_data['Organic Matter (%)'].mean()
    
    # Price premium calculation
    national_avg_price = national_results[national_results['Crop'] == crop]['Avg_Price'].mean()
    state_avg_price = state_crop_data['Wholesale Price'].mean()
    price_premium = ((state_avg_price - national_avg_price) / national_avg_price) * 100 if national_avg_price > 0 else 0
    
    # Generate dynamic rationale points
    rationale_points = []
    
    # Market position
    if market_share > 20:
        rationale_points.append(f"‚Ä¢ **Market Leader:** Accounts for {market_share:.1f}% of India's {crop} production")
    elif market_share > 10:
        rationale_points.append(f"‚Ä¢ **Significant Producer:** Contributes {market_share:.1f}% of national {crop} output")
    else:
        rationale_points.append(f"‚Ä¢ **Growing Producer:** Holds {market_share:.1f}% share in {crop} market")
    
    # Export potential
    if export_ratio > 30:
        rationale_points.append(f"‚Ä¢ **Export Powerhouse:** {export_ratio:.1f}% of production is exported globally")
    elif export_ratio > 15:
        rationale_points.append(f"‚Ä¢ **Growing Exporter:** {export_ratio:.1f}% export ratio shows international demand")
    else:
        rationale_points.append(f"‚Ä¢ **Domestic Focus:** Primarily serves local markets (export ratio: {export_ratio:.1f}%)")
    
    # Climate resilience
    if rainfall_var < 15:
        rationale_points.append(f"‚Ä¢ **Climate Stable:** Highly predictable rainfall (variability: {rainfall_var:.1f}%)")
    elif rainfall_var < 30:
        rationale_points.append(f"‚Ä¢ **Moderate Climate Risk:** Rainfall varies by {rainfall_var:.1f}% annually")
    else:
        rationale_points.append(f"‚Ä¢ **Climate Challenge:** High rainfall variability ({rainfall_var:.1f}%) requires irrigation planning")
    
    # Soil quality
    if soil_ph >= 6 and soil_ph <= 7 and organic_matter >= 2:
        rationale_points.append(f"‚Ä¢ **Premium Soil:** Ideal pH ({soil_ph:.1f}) and high organic content ({organic_matter:.1f}%)")
    elif (soil_ph >= 5.5 and soil_ph < 6) or (soil_ph > 7 and soil_ph <= 7.5):
        rationale_points.append(f"‚Ä¢ **Good Soil:** Moderate pH ({soil_ph:.1f}) and organic matter ({organic_matter:.1f}%)")
    else:
        rationale_points.append(f"‚Ä¢ **Soil Improvement Needed:** pH {soil_ph:.1f}, organic matter {organic_matter:.1f}%")
    
    # Price premium
    if price_premium > 10:
        rationale_points.append(f"‚Ä¢ **Premium Pricing:** Local prices are {price_premium:.1f}% above national average")
    elif price_premium > 0:
        rationale_points.append(f"‚Ä¢ **Slight Price Advantage:** State commands a {price_premium:.1f}% price premium")
    elif price_premium < -10:
        rationale_points.append(f"‚Ä¢ **Price Disadvantage:** Prices are {abs(price_premium):.1f}% below national average")
    else:
        rationale_points.append(f"‚Ä¢ **Competitive Pricing:** Prices align closely with national averages")
    
    # State-specific specializations
    if state == 'Punjab' and crop == 'Wheat':
        rationale_points.append("‚Ä¢ **Punjab Special:** Benefits from established wheat procurement system")
    elif state == 'Maharashtra' and crop == 'Grapes':
        rationale_points.append("‚Ä¢ **Maharashtra Advantage:** World-class grape processing infrastructure")
    
    return rationale_points

def analyze_all_states():
    all_results = []
    all_forecasts = {}
    available_states = merged_df['State'].unique()
    
    print("\n" + "="*100)
    print("COMPREHENSIVE AGRICULTURAL ANALYSIS ACROSS ALL INDIAN STATES (10-YEAR FORECAST)")
    print("="*100)
    print(f"\nAvailable states: {', '.join(available_states)}\n")
    
    # Analyze each state
    for state_name in available_states:
        state_df = merged_df[merged_df['State'] == state_name]
        state_results, state_forecasts = analyze_state(state_name, state_df)
        
        if state_results is not None:
            print_state_report(state_name, state_results, state_forecasts)
            all_results.append(state_results)
            all_forecasts[state_name] = state_forecasts
    
    if not all_results:
        print("\nNo valid data available for any state.")
        return None, None
    
    # Combine all results and find top 5 nationally
    national_results = pd.concat(all_results)
    national_results['National_Rank'] = national_results['Score'].rank(ascending=False)
    national_results = national_results.sort_values('National_Rank')
    top_5_national = national_results.head(5)
    
    # Print national recommendations
    print("\n" + "="*100)
    print("TOP 5 AGRICULTURAL INVESTMENT OPPORTUNITIES ACROSS INDIA (10-YEAR FORECAST)")
    print("="*100)
    
    for i, (_, row) in enumerate(top_5_national.iterrows(), 1):
        print(f"\nüèÜ RECOMMENDATION #{i}: {row['Crop'].upper()} in {row['State'].upper()}")
        print(f"üìç Current Production: {row['Current_Production']:,.2f} tonnes")
        print(f"üîÆ Forecasted Production (10yr avg): {row['Forecasted_Production']:,.2f} tonnes")
        print(f"üìà 10-Year Growth Rate: {row['Growth_Rate']:.2%}")
        print(f"üí∞ Average Price: ‚Çπ{row['Avg_Price']:,.2f}")
        print(f"üå± Soil Score: {row['Soil_Score']}/5")
        print(f"üèÜ 10-Year Investment Score: {row['Score']:,.2f}")

    # ===== DYNAMIC INVESTMENT RATIONALE =====
    print("\n" + "="*100)
    print("üí° DATA-DRIVEN INVESTMENT RATIONALE")
    print("="*100)
    
    for i, (_, row) in enumerate(top_5_national.iterrows(), 1):
        print(f"\nüîç **Why invest in {row['Crop']} in {row['State']}?**")
        rationale_points = generate_investment_rationale(row['State'], row['Crop'], merged_df, national_results)
        for point in rationale_points:
            print(point)
    
    # Plot top 5 national crops comparison
    plt.figure(figsize=(14, 8))
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    
    for i, (_, row) in enumerate(top_5_national.iterrows()):
        plt.bar(f"{row['Crop']}\n({row['State']})", row['Score'], color=colors[i], label=f"#{i+1}")
        
    plt.title("Top 5 Agricultural Investment Opportunities Across India\n(10-Year Forecast Period)", pad=20)
    plt.xlabel("Crop and State", labelpad=10)
    plt.ylabel("10-Year Investment Potential Score", labelpad=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title="Ranking")
    plt.tight_layout()
    plt.show()
    
    # Plot forecast graphs for each of the top 5 national crops
    print("\n" + "="*100)
    print("10-YEAR FORECAST GRAPHS FOR TOP 5 NATIONAL CROPS")
    print("="*100)
    
    for i, (_, row) in enumerate(top_5_national.iterrows(), 1):
        state_forecasts = all_forecasts[row['State']]
        if row['Crop'] in state_forecasts:
            print(f"\n10-Year Forecast for #{i}: {row['Crop']} in {row['State']}")
            plot_state_forecast(row['State'], row, state_forecasts)
    
    return national_results, all_forecasts

# Run the analysis
national_results, all_forecasts = analyze_all_states()

print("\nAnalysis complete. All state and national recommendations with 10-year forecasts are shown above!")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load and clean data
it_df = pd.read_csv(r"C:\Users\beraj\Downloads\IT_Sector_India_2010_2020.csv")
it_df.replace([np.inf, -np.inf], np.nan, inplace=True)
it_df.dropna(inplace=True)

def analyze_state(state_name, state_df, forecast_years=10):
    try:
        # Feature engineering
        state_df['Revenue_Growth'] = state_df['State_IT_Revenue(Cr)'].pct_change().fillna(0)
        
        # Fit model
        model = SARIMAX(
            state_df['State_IT_Revenue(Cr)'],
            exog=state_df[['Repo_Rate(%)', 'Global_Economic_Index']],
            order=(1, 1, 1),
            seasonal_order=(0, 0, 0, 0)
        )
        model_fit = model.fit(disp=False)
        
        # Generate forecast
        forecast = model_fit.get_forecast(
            steps=forecast_years,
            exog=pd.DataFrame({
                'Repo_Rate(%)': [state_df['Repo_Rate(%)'].iloc[-1]] * forecast_years,
                'Global_Economic_Index': [state_df['Global_Economic_Index'].iloc[-1]] * forecast_years
            })
        )
        
        return {
            'state': state_name,
            'current_revenue': state_df['State_IT_Revenue(Cr)'].iloc[-1],
            'avg_growth': state_df['Revenue_Growth'].mean(),
            'unemployment': state_df['Urban_Unemployment_Rate(%)'].iloc[-1],
            'internet_penetration': state_df['Internet_Penetration(%)'].iloc[-1],
            'historical_revenue': state_df['State_IT_Revenue(Cr)'].values,
            'forecast_revenue': forecast.predicted_mean.values,
            'conf_int': forecast.conf_int(),
            'historical_years': state_df['Year'].values,
            'forecast_years': [state_df['Year'].iloc[-1] + i + 1 for i in range(forecast_years)]
        }
    except Exception as e:
        print(f"Error processing {state_name}: {str(e)}")
        return None

def print_state_details(results):
    print("\n" + "="*80)
    print("DETAILED STATE-WISE IT SECTOR PERFORMANCE")
    print("="*80)
    
    for state in results:
        if state:  # Only process valid results
            growth = (state['forecast_revenue'].mean() - state['current_revenue']) / state['current_revenue']
            print(f"\nüìå {state['state'].upper()}")
            print(f"‚Ä¢ Current Revenue: ‚Çπ{state['current_revenue']:,.2f} Cr")
            print(f"‚Ä¢ Avg Historical Growth: {state['avg_growth']:.2%}")
            print(f"‚Ä¢ Urban Unemployment: {state['unemployment']:.2f}%")
            print(f"‚Ä¢ Internet Penetration: {state['internet_penetration']:.2f}%")
            print(f"‚Ä¢ 10-Year Projected Growth: {growth:.2%}")

def plot_combined_line_graph(results):
    plt.figure(figsize=(14, 7))
    
    for state in results:
        if state:
            all_years = list(state['historical_years']) + state['forecast_years']
            all_revenue = list(state['historical_revenue']) + list(state['forecast_revenue'])
            
            plt.plot(state['historical_years'], 
                     state['historical_revenue'], 
                     label=f"{state['state']} (Historical)")
            plt.plot(state['forecast_years'], 
                     state['forecast_revenue'], 
                     linestyle='--', 
                     label=f"{state['state']} (Forecast)")
    
    plt.title("IT Revenue Trends Across All States (10-Year Forecast)", pad=20)
    plt.xlabel("Year")
    plt.ylabel("Revenue (‚Çπ Cr)")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def plot_top3_bar_chart(top_states):
    plt.figure(figsize=(10, 6))
    
    states = [s['state'] for s in top_states if s]
    growth_rates = [
        (s['forecast_revenue'].mean() - s['current_revenue']) / s['current_revenue']
        for s in top_states if s
    ]
    
    if not states:
        print("No valid states for bar chart")
        return
    
    bars = plt.bar(states, growth_rates, color=['gold', 'silver', 'lightblue'])
    
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2., height,
                 f'{height:.2%}',
                 ha='center', va='bottom')
    
    plt.title("Top 3 States by Projected Growth Rate", pad=20)
    plt.ylabel("10-Year Growth Projection")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def generate_investment_strategy(top_states):
    print("\n" + "="*80)
    print("STRATEGIC INVESTMENT PLAN FOR TOP 3 STATES")
    print("="*80)
    
    # Compute national averages for comparison
    all_states = [s for s in top_states if s]
    national_unemployment = np.mean([s['unemployment'] for s in all_states])
    national_internet_penetration = np.mean([s['internet_penetration'] for s in all_states])
    national_growth = np.mean([
        (s['forecast_revenue'].mean() - s['current_revenue']) / s['current_revenue']
        for s in all_states
    ])
    
    for i, state in enumerate(all_states[:3], 1):
        growth_rate = (
            (state['forecast_revenue'].mean() - state['current_revenue']) / state['current_revenue']
        )
        revenue_volatility = np.std(state['historical_revenue']) / np.mean(state['historical_revenue'])

        print(f"\nüèÜ #{i}: {state['state'].upper()} (Projected Growth: {growth_rate:.2%})")
        
        # Strategic Reasons
        print("\nWHY INVEST HERE?")
        print(f"‚Ä¢ Growth rate ({growth_rate:.2%}) {'exceeds' if growth_rate > national_growth else 'is below'} the national average ({national_growth:.2%})")
        print(f"‚Ä¢ Urban Unemployment Rate: {state['unemployment']:.2f}% "
              f"({'Lower' if state['unemployment'] < national_unemployment else 'Higher'} than national average: {national_unemployment:.2f}%)")
        print(f"‚Ä¢ Internet Penetration: {state['internet_penetration']:.2f}% "
              f"({'Above' if state['internet_penetration'] > national_internet_penetration else 'Below'} national average: {national_internet_penetration:.2f}%)")
        print(f"‚Ä¢ Revenue Volatility: {revenue_volatility:.2%} (Lower is better)")
        
        # Forecast Confidence
        conf_int = state['conf_int']
        conf_width = (conf_int.iloc[:, 1] - conf_int.iloc[:, 0]).mean()
        print(f"‚Ä¢ Forecast Confidence Interval (avg. width): ¬±‚Çπ{conf_width:.2f} Cr ‚Äî "
              f"{'Stable' if conf_width < state['current_revenue'] * 0.15 else 'Volatile'} outlook")

        print("\nCORE ALLOCATION:")
        print(f"‚Ä¢ 50% to established IT firms (Current Revenue: ‚Çπ{state['current_revenue']:,.2f} Cr)")
        print(f"‚Ä¢ 30% to digital infrastructure (Internet Penetration: {state['internet_penetration']:.2f}%)")
        print(f"‚Ä¢ 20% to workforce development (Unemployment: {state['unemployment']:.2f}%)")
        
        print("\nSPECIAL INITIATIVES:")
        if growth_rate > 0.2:
            print("- Create special economic zone for tech companies")
            print("- Offer 5-year tax holiday for new IT investments")
        elif growth_rate > 0.1:
            print("- Upgrade existing IT parks with 5G infrastructure")
            print("- Subsidize tech education programs")
        else:
            print("- Implement business retention grants")
            print("- Develop regional innovation hubs")


# Main analysis
def run_analysis():
    all_results = []
    for state_name in it_df['State'].unique():
        state_data = it_df[it_df['State'] == state_name].sort_values('Year')
        result = analyze_state(state_name, state_data)
        all_results.append(result)
    
    print_state_details(all_results)
    
    valid_results = [r for r in all_results if r is not None]
    valid_results.sort(key=lambda x: ((x['forecast_revenue'].mean() - x['current_revenue']) / x['current_revenue']), reverse=True)
    top3 = valid_results[:3]
    
    plot_combined_line_graph(valid_results)
    plot_top3_bar_chart(top3)
    generate_investment_strategy(top3)

# Execute analysis
print("INDIAN IT SECTOR INVESTMENT ANALYSIS")
print("="*80)
run_analysis()
print("\nAnalysis completed successfully!")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load and clean data
it_df = pd.read_csv(r"D:\Projects\GDP\data\raw\IT_Sector_India_2010_2020.csv")
it_df.replace([np.inf, -np.inf], np.nan, inplace=True)
it_df.dropna(inplace=True)

def analyze_state(state_name, state_df, forecast_years=10):
    try:
        # Feature engineering
        state_df['Revenue_Growth'] = state_df['State_IT_Revenue(Cr)'].pct_change().fillna(0)
        
        # Fit model
        model = SARIMAX(
            state_df['State_IT_Revenue(Cr)'],
            exog=state_df[['Repo_Rate(%)', 'Global_Economic_Index']],
            order=(1, 1, 1),
            seasonal_order=(0, 0, 0, 0)
        )
        model_fit = model.fit(disp=False)
        
        # Generate forecast
        forecast = model_fit.get_forecast(
            steps=forecast_years,
            exog=pd.DataFrame({
                'Repo_Rate(%)': [state_df['Repo_Rate(%)'].iloc[-1]] * forecast_years,
                'Global_Economic_Index': [state_df['Global_Economic_Index'].iloc[-1]] * forecast_years
            })
        )
        
        return {
            'state': state_name,
            'current_revenue': state_df['State_IT_Revenue(Cr)'].iloc[-1],
            'avg_growth': state_df['Revenue_Growth'].mean(),
            'unemployment': state_df['Urban_Unemployment_Rate(%)'].iloc[-1],
            'internet_penetration': state_df['Internet_Penetration(%)'].iloc[-1],
            'historical_revenue': state_df['State_IT_Revenue(Cr)'].values,
            'forecast_revenue': forecast.predicted_mean.values,
            'conf_int': forecast.conf_int(),
            'historical_years': state_df['Year'].values,
            'forecast_years': [state_df['Year'].iloc[-1] + i + 1 for i in range(forecast_years)]
        }
    except Exception as e:
        print(f"Error processing {state_name}: {str(e)}")
        return None

def print_state_details(results):
    print("\n" + "="*80)
    print("DETAILED STATE-WISE IT SECTOR PERFORMANCE")
    print("="*80)
    
    for state in results:
        if state:  # Only process valid results
            growth = (state['forecast_revenue'].mean() - state['current_revenue']) / state['current_revenue']
            print(f"\nüìå {state['state'].upper()}")
            print(f"‚Ä¢ Current Revenue: ‚Çπ{state['current_revenue']:,.2f} Cr")
            print(f"‚Ä¢ Avg Historical Growth: {state['avg_growth']:.2%}")
            print(f"‚Ä¢ Urban Unemployment: {state['unemployment']:.2f}%")
            print(f"‚Ä¢ Internet Penetration: {state['internet_penetration']:.2f}%")
            print(f"‚Ä¢ 10-Year Projected Growth: {growth:.2%}")

def plot_individual_state_graphs(results):
    for state in results:
        if state:
            plt.figure(figsize=(10, 6))
            
            # Historical data
            plt.plot(state['historical_years'], 
                     state['historical_revenue'], 
                     'b-', linewidth=2, 
                     label='Historical Revenue')
            
            # Forecast data
            plt.plot(state['forecast_years'], 
                     state['forecast_revenue'], 
                     'r--', linewidth=2, 
                     label='Forecast Revenue')
            
            # Confidence interval
            conf_int = state['conf_int']
            plt.fill_between(state['forecast_years'],
                            conf_int.iloc[:, 0],
                            conf_int.iloc[:, 1],
                            color='pink', alpha=0.3,
                            label='Confidence Interval')
            
            plt.title(f"IT Revenue Trend for {state['state']} (10-Year Forecast)", pad=20)
            plt.xlabel("Year")
            plt.ylabel("Revenue (‚Çπ Cr)")
            plt.legend()
            plt.grid(True, linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()

def plot_combined_line_graph(results):
    plt.figure(figsize=(14, 7))
    
    for state in results:
        if state:
            all_years = list(state['historical_years']) + state['forecast_years']
            all_revenue = list(state['historical_revenue']) + list(state['forecast_revenue'])
            
            plt.plot(state['historical_years'], 
                     state['historical_revenue'], 
                     label=f"{state['state']} (Historical)")
            plt.plot(state['forecast_years'], 
                     state['forecast_revenue'], 
                     linestyle='--', 
                     label=f"{state['state']} (Forecast)")
    
    plt.title("IT Revenue Trends Across All States (10-Year Forecast)", pad=20)
    plt.xlabel("Year")
    plt.ylabel("Revenue (‚Çπ Cr)")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def plot_top3_bar_chart(top_states):
    plt.figure(figsize=(10, 6))
    
    states = [s['state'] for s in top_states if s]
    growth_rates = [
        (s['forecast_revenue'].mean() - s['current_revenue']) / s['current_revenue']
        for s in top_states if s
    ]
    
    if not states:
        print("No valid states for bar chart")
        return
    
    colors = ['gold', 'silver', '#cd7f32']  # Gold, Silver, Bronze colors
    bars = plt.bar(states, growth_rates, color=colors)
    
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2., height,
                 f'{height:.2%}',
                 ha='center', va='bottom')
    
    plt.title("Top 3 States by Projected Growth Rate", pad=20)
    plt.ylabel("10-Year Growth Projection")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def highlight_top3_states(top_states):
    print("\n" + "="*80)
    print("üèÜ TOP 3 STATES FOR IT INVESTMENT üèÜ")
    print("="*80)
    
    for i, state in enumerate(top_states[:3], 1):
        growth_rate = ((state['forecast_revenue'].mean() - state['current_revenue']) / 
                      state['current_revenue'])
        
        print(f"\nü•á #{i}: {state['state'].upper()}")
        print(f"   Projected 10-Year Growth: {growth_rate:.2%}")
        print(f"   Current IT Revenue: ‚Çπ{state['current_revenue']:,.2f} Cr")
        print(f"   Key Advantages:")
        print(f"   ‚Ä¢ Urban Unemployment: {state['unemployment']:.2f}% (Below national average)")
        print(f"   ‚Ä¢ Internet Penetration: {state['internet_penetration']:.2f}% (Above national average)")
        print(f"   ‚Ä¢ Historical Growth Stability: {state['avg_growth']:.2%}")

def generate_investment_strategy(top_states):
    print("\n" + "="*80)
    print("STRATEGIC INVESTMENT PLAN FOR TOP 3 STATES")
    print("="*80)
    
    # Compute national averages for comparison
    all_states = [s for s in top_states if s]
    national_unemployment = np.mean([s['unemployment'] for s in all_states])
    national_internet_penetration = np.mean([s['internet_penetration'] for s in all_states])
    national_growth = np.mean([
        (s['forecast_revenue'].mean() - s['current_revenue']) / s['current_revenue']
        for s in all_states
    ])
    
    for i, state in enumerate(all_states[:3], 1):
        growth_rate = (
            (state['forecast_revenue'].mean() - state['current_revenue']) / state['current_revenue']
        )
        revenue_volatility = np.std(state['historical_revenue']) / np.mean(state['historical_revenue'])

        print(f"\nüèÜ #{i}: {state['state'].upper()} (Projected Growth: {growth_rate:.2%})")
        
        # Strategic Reasons
        print("\nWHY INVEST HERE?")
        print(f"‚Ä¢ Growth rate ({growth_rate:.2%}) {'exceeds' if growth_rate > national_growth else 'is below'} the national average ({national_growth:.2%})")
        print(f"‚Ä¢ Urban Unemployment Rate: {state['unemployment']:.2f}% "
              f"({'Lower' if state['unemployment'] < national_unemployment else 'Higher'} than national average: {national_unemployment:.2f}%)")
        print(f"‚Ä¢ Internet Penetration: {state['internet_penetration']:.2f}% "
              f"({'Above' if state['internet_penetration'] > national_internet_penetration else 'Below'} national average: {national_internet_penetration:.2f}%)")
        print(f"‚Ä¢ Revenue Volatility: {revenue_volatility:.2%} (Lower is better)")
        
        # Forecast Confidence
        conf_int = state['conf_int']
        conf_width = (conf_int.iloc[:, 1] - conf_int.iloc[:, 0]).mean()
        print(f"‚Ä¢ Forecast Confidence Interval (avg. width): ¬±‚Çπ{conf_width:.2f} Cr ‚Äî "
              f"{'Stable' if conf_width < state['current_revenue'] * 0.15 else 'Volatile'} outlook")

        print("\nCORE ALLOCATION:")
        print(f"‚Ä¢ 50% to established IT firms (Current Revenue: ‚Çπ{state['current_revenue']:,.2f} Cr)")
        print(f"‚Ä¢ 30% to digital infrastructure (Internet Penetration: {state['internet_penetration']:.2f}%)")
        print(f"‚Ä¢ 20% to workforce development (Unemployment: {state['unemployment']:.2f}%)")
        
        print("\nSPECIAL INITIATIVES:")
        if growth_rate > 0.2:
            print("- Create special economic zone for tech companies")
            print("- Offer 5-year tax holiday for new IT investments")
        elif growth_rate > 0.1:
            print("- Upgrade existing IT parks with 5G infrastructure")
            print("- Subsidize tech education programs")
        else:
            print("- Implement business retention grants")
            print("- Develop regional innovation hubs")

# Main analysis
def run_analysis():
    all_results = []
    for state_name in it_df['State'].unique():
        state_data = it_df[it_df['State'] == state_name].sort_values('Year')
        result = analyze_state(state_name, state_data)
        all_results.append(result)
    
    print_state_details(all_results)
    
    valid_results = [r for r in all_results if r is not None]
    valid_results.sort(key=lambda x: ((x['forecast_revenue'].mean() - x['current_revenue']) / x['current_revenue']), reverse=True)
    top3 = valid_results[:3]
    
    # Plot individual state graphs
    plot_individual_state_graphs(valid_results)
    
    # Plot combined and top 3 graphs
    plot_combined_line_graph(valid_results)
    plot_top3_bar_chart(top3)
    
    # Highlight top 3 states
    highlight_top3_states(top3)
    
    # Generate investment strategy
    generate_investment_strategy(top3)

# Execute analysis
print("INDIAN IT SECTOR INVESTMENT ANALYSIS")
print("="*80)
run_analysis()
print("\nAnalysis completed successfully!")
```
